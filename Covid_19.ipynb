```python
from scipy.integrate import odeint as od
import numpy as np
import seaborn as sns ; sns.set()
import  matplotlib.pyplot as plt
from IPython.display import HTML
from ipywidgets.widgets import interact, IntSlider, FloatSlider, Layout
%matplotlib inline
plt.style.use('fivethirtyeight')
style = {'description_width': '150px'}
slider_layout = Layout(width='99%')
```

```python
def corona(Istart,trans_rate,reco_rate,population,days):
    Sstart=.99
    Rstart=0
    t=np.linspace(1,days,days)
    
    
    def sol(c,t):
        s,i,r=c
        S=-s*i*trans_rate
        I=trans_rate*s*i-reco_rate*i
        R=reco_rate*i
        
        return [S,I,R]
    
    def new_cases(y,p,d):
        y_=np.diff(y,1)*p
        y_[y_<0]=0
        x_=np.linspace(1,d-1,d-1)
        return x_,y_
    
    Sol=od(sol,[Sstart,Istart,Rstart],t)
    
    Suspect  = Sol[:,0]
    Infected = Sol[:,1]
    Recoverd = Sol[:,2]
    
    fig,ax=plt.subplots(1,2,figsize=(15,7),sharey=False)
    fig.tight_layout(pad=5)
    
    ax[0].plot(Suspect  ,c='r',label='Suspect')
    ax[0].plot(Infected,c='b',label='Infected')
    ax[0].plot(Recoverd,c='g',label='Recoverd')
    ax[0].legend(loc='best')
    ax[0].set(xlabel='Days',ylabel='Population %',title='Covid-19 SIR Model')
    
    I_x,I_y=new_cases(Infected,population,days)
    S_x,S_y=I_x,Suspect[:-1]*population
    R_x,R_y=new_cases(Recoverd,population,days)
    
    ax[1].bar(S_x,S_y,color='r',label='Suspicious'      ,alpha=0.5)
    ax[1].bar(I_x,I_y,color='b',label='New cases',alpha=0.5)
    ax[1].bar(R_x,R_y,color='g',label='Recovered People',alpha=0.5)
    ax[1].legend(loc='best')
    ax[1].set(xlabel='Days',ylabel='Number of people',title='Effect to the commiunity - {} People'.format(population))
    
    return  
    
```

```python
interact(corona,Istart=FloatSlider(min=0.01, max=1, step=0.05 ,value=0.01, description='Initial infected ppl (%)'              ,style=style, layout=slider_layout     ),
             trans_rate=FloatSlider(min=0, max=10, step=0.01, value=2.5, description='Transfer Rate' ,style=style, layout=slider_layout        ),
             reco_rate=FloatSlider(min=0, max=1, step=0.01, value=.2, description='Recover Rate'   ,style=style, layout=slider_layout      ),
             days=IntSlider(min=1, max=100, step=1, value=21, description='Days'                    ,style=style, layout=slider_layout    ),
            population=IntSlider(min=1, max=10000, step=1, value=21, description='Population size'                    ,style=style, layout=slider_layout    ));
```
